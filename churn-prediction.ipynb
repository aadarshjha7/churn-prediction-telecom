```python
columns1=['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
```

```python
for i in columns1:
    if i == 'gender':
        df[i] = df[i].map({'Male': 1, 'Female': 0})
    else:
        df[i] = df[i].map({'Yes': 1, 'No': 0})
        
```

```python
df
```

```python
columns2= ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract','PaymentMethod']
```

```python
df_transformed = pd.get_dummies(df, columns=columns2)
```

```python
df.head()
```

```python
df_transformed.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
df1 = df_transformed.copy()
```

```python
for i in df_transformed.columns:
    df1[i] = label_encoder.fit_transform(df_transformed[i])
```

```python
df1.head()
```

## Logistic Regression

```python
X = df1.drop('Churn', axis=1)
Y = df1['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
accuracy = metrics.accuracy_score(y_test, pred)
```
